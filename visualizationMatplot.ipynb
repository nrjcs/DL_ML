# **visualizing with help of graphs: Matplotlib**

```
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## **method for Graph Plotting **

```
def plot(xaxis, yaxis, xlabel="x axis", xfont=15, ylabel="y axis", yfont=15, title="title", tfont=15):
    
    ##ytics calibration
    yticks=[x for x in range(int(np.amax(yaxis))+2)] 
    #yticks=[x for x in range(max(yaxis)+2)] #for list
    xticks=[x for x in range(int(np.amax(xaxis))+2)] 
    #xticks=[x for x in range(max(xaxis)+2)] #for list
    
    ##ploting line or bar
    plt.bar(xaxis, yaxis) #bar graph
    #plt.plot(xaxis, yaxis) #line
    #plt.plot(xaxis, yaxis, "ro") #dot
    
    
    plt.xlabel(xlabel, fontsize=xfont)
    plt.ylabel(ylabel, fontsize=yfont)
    plt.xticks(xticks, fontsize=15, rotation=30)
    plt.yticks(yticks, fontsize=15, rotation=30)
    plt.title(title, fontsize=tfont)
    plt.show()
    return
```

## **Plotting Line and Bar graph with Numpy array as X and Y**

# **Creating Pandas Dataframe with Dictionary**

```
df_Dictionary=pd.DataFrame({"Name":["John", "Smith", "Bob", "Ted"], "Subject":["Maths", "Science", "English", "Economics"], "Percent marks":[80, 85, 78, 90], "Class":[6, 8, 4, 9]})
```

## **Creating Pandas Dataframe with List**

```
df=pd.DataFrame(columns=["Name", "Subject", "Percent marks", "Class"])
df["Name"]=["John", "Smith", "Bob", "Ted"]
df["subject"]=["Maths", "Science", "English", "Economics"]
df["Percent marks"]=[80, 85, 78, 90]
df["Class"]=[6, 8, 4, 9]
```

## **Dataframe Condition and Selection**

```
print df[( (df["Percent marks"]==df["Percent marks"].max()) )].iloc[0]["Percent marks"]
class_List=list(df["Class"])
print class_List
print "average of the class:", np.mean(class_List)
print "maximum of the class:", max(class_List)
```

# **Plotting Line and Bar graphs with Dataframe  **

```
df.plot(x="Name", y=["Percent marks"], linewidth=4, fontsize=30, xticks=[x for x in range(df.shape[0])])#line plot
df.plot(x="Name", y=["Percent marks", "Class"], linewidth=4, fontsize=30, xticks=[x for x in range(df.shape[0])])#multi line plot
df.plot(x="Name", y=["Percent marks"], kind="bar", fontsize=30) #bar plot
df.plot(x="Name", y=["Percent marks", "Class"], kind="bar", fontsize=30) #multi-bar plot
```
